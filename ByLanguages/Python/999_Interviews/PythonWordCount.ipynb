# Python Word Count Program

```python
def WordCount(filename, mode):
    file = open(filename, mode)
    wordcount = {}
    for word in file.read().split():
        if word not in wordcount:
            wordcount[word] = 1
        else:
            wordcount[word] += 1
    for key in sorted(wordcount.keys()):
        print(key, wordcount[key])
    file.close();
```

```python
WordCount("TestData.txt","r+")
```

```python
WordCount("HarryPotter.txt","r+")
```
